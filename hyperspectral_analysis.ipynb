```python
# Importing Standard cyan
hdr_standard = 'cyan-standard'
img_standard = envi.open('hyperspectral_files/' + hdr_standard + '.hdr')

# Importing Fluorescent cyan
hdr_fluorescent = 'cyan-fluorescent'
img_fluorescent = envi.open('hyperspectral_files/' + hdr_fluorescent + '.hdr')
```

```python
# Simulate R, G, B view
R = int(np.argmin(np.abs(np.array(img_standard.bands.centers) - 700)))
G = int(np.argmin(np.abs(np.array(img_standard.bands.centers) - 546.1)))
B = int(np.argmin(np.abs(np.array(img_standard.bands.centers) - 435.8)))

# Display RGB image (standard)
view_rgb = imshow(img_standard, (R, G, B))
```

```python
# Display only one band (standard)
# Band 770 equals to ≈ 950 nm
view_band = imshow(img_standard[40:959, 190:1109, 770])
```
